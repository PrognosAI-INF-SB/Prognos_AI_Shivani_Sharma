```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# File path
filepath = "sensor_data.csv"

# Load CSV
df = pd.read_csv(filepath)
print("✅ File loaded successfully!")
print("Columns:", df.columns.tolist())

# Fill missing values
df.fillna(method='ffill', inplace=True)

# Standardize sensors
sensor_cols = ['SensorA', 'SensorB', 'SensorC']
scaler = StandardScaler()
df[sensor_cols] = scaler.fit_transform(df[sensor_cols])

# Compute RUL
df['RUL'] = np.arange(len(df)-1, -1, -1)

# Create rolling windows
def create_rolling_windows(df, window_size=30):
    sequences = []
    for i in range(len(df)-window_size):
        sequences.append(df.iloc[i:i+window_size].values)
    return np.array(sequences)

sequences = create_rolling_windows(df, window_size=30)

# Output
print("\n✅ Processed data shape:", df.shape)
print("✅ Sequences shape:", sequences.shape)
print("✅ Sample RUL values:\n", df['RUL'].head())
```
